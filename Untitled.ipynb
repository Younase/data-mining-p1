```python
# rawdata=df['date'].copy()
datan=dict()
datas=dict()
datan['date']=sorted(list(set(df['date'])))#df['date'].copy()
datas['date']=sorted(list(set(df['date'])))#df['date'].copy()
l=len(datan['date'])
for n in niches:
    datan[n]=[0]*l
for s in subniches:
    datas[s]=[0]*l
# print(datan)
# print(subniches)
# print(datas)
i=0
j=0
for d,date in zip(df['description'],df['date']):
#     d=df['description'][0]
    descTechs=dict()
    descN=dict()
    descS=dict()
    desc=d.lower()
    desc=tokenizer.tokenize(desc)
    
    for e in desc:
        if e in techs:
            try:
                    descTechs[e]+=1
            except KeyError:
                    descTechs[e]=1
    for t in descTechs:
        try:
            descS[map[t]]+=descTechs[t]
        except KeyError:
            descS[map[t]]=descTechs[t]
    for s in descS:
        try:
            descN[nichemap[s]]+=descS[s]
        except KeyError:
            descN[nichemap[s]]=descS[s]

    if len(descTechs)==0:
        continue
    idx=datan['date'].index(date)
    try:
        datas[max(descS, key=descS.get)][idx]+=1
        datan[max(descN, key=descN.get)][idx]+=1
    except:
        pass
```
